```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "I love machine learning",
    "Deep learning is powerful",
    "Python is great for data science",
    "Neural networks can classify images",
    "Natural language processing is amazing",
    "I enjoy coding in Python",
    "Data analysis helps in decision making",
    "Artificial intelligence is the future",
    "TensorFlow and PyTorch are popular frameworks",
    "Big data requires efficient algorithms"
]
```

```python
sent
```

```python
#vocabulary size
voc_size=10000
```

```python
#One_hot
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=7
emb_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(emb_docs)
```

```python
#feature representaion
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim))
model.build(input_shape=(None, sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(emb_docs)
```
